```python
import requests
from bs4 import BeautifulSoup

# r = requests.get("https://www.century21.com/real-estate/rock-wv/LCWVROCK/", headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})
r = requests.get("https://www.century21.com/real-estate/califon-nj/LCNJCALIFON/", headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})

c = r.content

soup = BeautifulSoup(c, "html.parser")

all = soup.find_all("div", {"class":"infinite-item"})


data_list = []

for item in all:
    if item.find("a", {"class": "listing-price"}):   
        data_dict = {}
        try:
            price =  item.find("a", {"class": "listing-price"}).text.replace("\n","").replace(" ","")
        except:
            price = None
        try:
            address =  item.find("div", {"class":"property-address"}).text.replace("\n","").replace("  ","")
        except:
            address = None
        try:
            city =  item.find("div", {"class": "property-city"}).text.replace("\n","").replace("  ","")
        except:
            city = None
        
        try:
            beds =  item.find("div", {"class": "property-beds"}).text.replace("\n","").replace("  ", "")
        except:
            beds = None
        try:
            baths = item.find("div", {"class": "property-baths"}).text.replace("\n","").replace("  ", "")
        except:
            baths=None
        try:
            property_size =item.find("div", {"class": "property-size"}).text.replace("\n","").replace("  ", "")
        except:
            property_size = None
        data_dict = {
            "price": price,
            "address" : address,
            "city" : city,
            "beds" : beds,
            "baths": baths,
            "property_size" : property_size           
            
        }
    data_list.append(data_dict)
        
        

import pandas as pd
df = pd.DataFrame(data_list)

df.to_csv("houseing1.csv")
df
```
